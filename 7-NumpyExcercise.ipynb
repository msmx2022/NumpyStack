```python
import numpy as np
from datetime import datetime
```

```python
# generate matrix 

M_1 = np.random.randn(100,2)
M_2 = np.random.randn(2,1)

```

```python
# set repeat times
T = 100000
```

```python
# generate list 

List_1 = []
for i in range(M_1.shape[0]):
    List_1.append(M_1[i].tolist())

List_2 = []
for i in range(M_2.shape[0]):
    List_2.append(M_2[i].tolist())
```

```python
# matrix multiplication

t0 = datetime.now()

for t in range(T):
    M_result = M_1.dot(M_2)

t1 = datetime.now() - t0
t1.total_seconds()

```

```python
# list multiplication

t0 = datetime.now()

for t in range(T):
    list_result = []
    for i in range(len(List_1)):
        dot = 0
        for j in range(len(List_1[i])):
            dot += List_1[i][j]*List_2[j][0]
        list_result.append(dot)


t2 = datetime.now() - t0
t2.total_seconds()
```

```python
# comparision

print("t2/t1 :",t2/t1)
```

```python
# test whether the result is same

list_sum = 0
for i in range(len(list_result)):
    list_sum += list_result[i]

M_sum = M_result.sum()

print("list_sum - M_sum = ",list_sum - M_sum)
np.allclose(list_sum,M_sum)
```
