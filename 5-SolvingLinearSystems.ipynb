# summary

1.inverse is slower and less accurate (np.linalg.inv -- no)

2.there are better algo to solve linear systems (np.linalg.solve -- yes)

```python
import numpy as np
```

```python
A = np.array([[1,1],[1.5,4]])
b = np.array([2200,5050])
```

```python
np.linalg.solve(A,b)
```

```python
# because data is small and easy is compute, so there is no difference
np.linalg.inv(A).dot(b)
```
